```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
plt.style.use("asludds")

gain_data = pd.read_csv("TIA_gain.csv").to_numpy()
noise_data = pd.read_csv("TIA_currentnoise.csv").to_numpy()
```

```python
plt.figure()
plt.semilogx(gain_data[:,0],gain_data[:,1])
plt.xlabel("Frequency (Hz)")
plt.ylabel("Gain (dB" + r"$\Omega$)")
plt.ylim(10,82)
plt.show()
```

```python
plt.figure()
plt.semilogx(noise_data[:,0],noise_data[:,1]*1e12)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Current noise density (pA/" + r"$\sqrt{Hz}$)")
plt.xlim(1e7,1e11)
plt.ylim(1,10)
plt.show()
```
