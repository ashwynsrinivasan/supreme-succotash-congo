```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
# plt.style.use("plot_style.mplstyle")
```

```python
psrnit_mc = pd.read_csv("gf_psrnit_loss_monte_carlo_glob_v1042a_20240416.csv")
```

## Polarization Controller

```python
psrnit_mc.columns
```

```python
plt.figure(figsize=(7.5,5))
for idd_temperature, temperature in enumerate([10, 25, 105]):
  plt.subplot(211)
  psrnit_mc_temp = psrnit_mc[psrnit_mc['temperature'] == temperature]
  plt.hist(-psrnit_mc_temp[f'fTEloss'] + 0.2, alpha=1.0, label=f'{temperature}C', bins=51, density = True)
  plt.legend()
  plt.title(f"TE to TE loss")
  plt.xlim([0,1.2])
  plt.ylim([0, 6])
  plt.grid(True)
  plt.subplot(212)
  plt.hist(-psrnit_mc_temp[f'fTMloss'] + 0.2, alpha=1.0, label=f'{temperature}C', bins=51, density = True)
  plt.legend()
  plt.title(f"TM to TE loss")
  plt.xlabel("IL [dB]")
  plt.xlim([0,1.2])
  plt.ylim([0,6])
  plt.grid(True)
  plt.tight_layout()
```

```python
psrnit_mc_temp = psrnit_mc[(psrnit_mc['temperature'] == 10)]
print(psrnit_mc_temp[f'fTMloss'].median(), psrnit_mc_temp[f'fTEloss'].median(), psrnit_mc_temp[f'fTMloss'].std(), psrnit_mc_temp[f'fTEloss'].std())
```
