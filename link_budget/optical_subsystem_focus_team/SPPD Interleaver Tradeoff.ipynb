```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
il_mc_20G = pd.read_csv('ILmc_20G_25C.csv')
no_bins = 51
jb = stats.jarque_bera(il_mc_20G[' IL (1 25C)'])
sk = stats.skew(il_mc_20G[' IL (1 25C)'])
kt = stats.kurtosis(il_mc_20G[' IL (1 25C)'])
median = np.median(il_mc_20G[' IL (1 25C)'])
std = np.std(il_mc_20G[' IL (1 25C)'])
plt.figure(figsize=(12, 6))
plt.subplot(221)
plt.hist(il_mc_20G[' IL (1 25C)'], bins=no_bins, density=True, label=r'Jarque-Bera = %.2f'%(jb[0])+ f'\n'+r'Skew = %.2f'%(sk) + f'\n' + r'Kurtosis = %.2f'%(kt) + f'\n' + r'Median = %.2f dB'%(median) + f'\n' + r'1-$\sigma$ = %.2f dB' %(std))
plt.legend()
plt.grid(True)
plt.title('Ch-1')
plt.xlim([-1, 0])
plt.xlabel('IL [dB]')

jb = stats.jarque_bera(il_mc_20G[' IL (2 25C)'])
sk = stats.skew(il_mc_20G[' IL (2 25C)'])
kt = stats.kurtosis(il_mc_20G[' IL (2 25C)'])
median = np.median(il_mc_20G[' IL (2 25C)'])
std = np.std(il_mc_20G[' IL (2 25C)'])

plt.subplot(222)
plt.hist(il_mc_20G[' IL (2 25C)'], bins=no_bins, density=True, label=r'Jarque-Bera = %.2f'%(jb[0])+ f'\n'+r'Skew = %.2f'%(sk) + f'\n' + r'Kurtosis = %.2f'%(kt) + f'\n' + r'Median = %.2f dB'%(median) + f'\n' + r'1-$\sigma$ = %.2f dB' %(std))
plt.legend()
plt.grid(True)
plt.title('Ch-4')
plt.xlim([-1, 0])
plt.xlabel('IL [dB]')

jb = stats.jarque_bera(il_mc_20G[' IL (3 25C)'])
sk = stats.skew(il_mc_20G[' IL (3 25C)'])
kt = stats.kurtosis(il_mc_20G[' IL (3 25C)'])
median = np.median(il_mc_20G[' IL (3 25C)'])
std = np.std(il_mc_20G[' IL (3 25C)'])

plt.subplot(223)
plt.hist(il_mc_20G[' IL (3 25C)'], bins=no_bins, density=True, label=r'Jarque-Bera = %.2f'%(jb[0])+ f'\n'+r'Skew = %.2f'%(sk) + f'\n' + r'Kurtosis = %.2f'%(kt) + f'\n' + r'Median = %.2f dB'%(median) + f'\n' + r'1-$\sigma$ = %.2f dB' %(std))
plt.legend()
plt.grid(True)
plt.title('Ch-5')
plt.xlim([-1, 0])
plt.xlabel('IL [dB]')

jb = stats.jarque_bera(il_mc_20G[' IL (4 25C)'])
sk = stats.skew(il_mc_20G[' IL (4 25C)'])
kt = stats.kurtosis(il_mc_20G[' IL (4 25C)'])
median = np.median(il_mc_20G[' IL (4 25C)'])
std = np.std(il_mc_20G[' IL (4 25C)'])

plt.subplot(224)
plt.hist(il_mc_20G[' IL (4 25C)'], bins=no_bins, density=True, label=r'Jarque-Bera = %.2f'%(jb[0])+ f'\n'+r'Skew = %.2f'%(sk) + f'\n' + r'Kurtosis = %.2f'%(kt) + f'\n' + r'Median = %.2f dB'%(median) + f'\n' + r'1-$\sigma$ = %.2f dB' %(std))
plt.legend()
plt.grid(True)
plt.title('Ch-8')
plt.xlim([-1, 0])
plt.xlabel('IL [dB]')
plt.tight_layout()
```
