```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
plt.style.use("plot_style.mplstyle")
```

```python
te_iosmfnit_connector = 0.65
tm_iosmfnit_connector = 0.72

connector_fiber = 0.7

te_iosmnfnit = 0.2
tm_iosmnfnit = 0.4
```

```python
std_te_iosmfnit_connector_fiber = 1.05
std_tm_iosmfnit_connector_fiber = 1.21

std_te_iosmnfnit = 0.1
std_tm_iosmnfnit = 0.1
```

```python
te_loss = te_iosmfnit_connector + connector_fiber + te_iosmnfnit
tm_loss = tm_iosmfnit_connector + connector_fiber + tm_iosmnfnit

std_te_loss = np.sqrt(std_te_iosmfnit_connector_fiber**2 + std_te_iosmnfnit**2)
std_tm_loss = np.sqrt(std_tm_iosmfnit_connector_fiber**2 + std_tm_iosmnfnit**2)
```

```python
labels = ['iosmnfnit','iosmfnit to\nconnector', 'connector to\nfiber']
```

```python

plt.figure()
plt.bar(np.arange(3)-0.125, [te_iosmnfnit, te_iosmfnit_connector, connector_fiber],width=0.25, label=f'TE:{te_loss:0.1f} [dB]')
plt.bar(np.arange(3)+0.125, [tm_iosmnfnit, tm_iosmfnit_connector, connector_fiber],width=0.25, label=f'TM:{tm_loss:0.1f} [dB]')
plt.ylabel('Mean Loss [dB]')
plt.xticks(np.arange(3), labels)
plt.grid(False)
plt.title('Single fiber attach')
plt.legend()
```

```python
labels = ['iosmnfnit','iosmfnit to\nfiber']
```

```python
plt.figure()
plt.bar(np.arange(2)-0.125, [te_iosmnfnit + std_te_iosmnfnit, te_iosmfnit_connector + connector_fiber + std_te_iosmfnit_connector_fiber],width=0.25, label=f'TE:{te_loss:0.1f} [dB]')
plt.bar(np.arange(2)+0.125, [tm_iosmnfnit + std_tm_iosmnfnit, te_iosmfnit_connector + connector_fiber + std_tm_iosmfnit_connector_fiber],width=0.25, label=f'TM:{tm_loss:0.1f} [dB]')
plt.ylabel(r'3-$\sigma$ Loss [dB]')
plt.xticks(np.arange(2), labels)
plt.grid(False)
plt.title('Single fiber attach')
plt.legend()
```
