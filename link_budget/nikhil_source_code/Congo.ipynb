```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

######LOAD IN DATA#########

# Load data from the Excel file into a Pandas DataFrame
file_path = 'FiberConn80_80_IL_non_elite.xlsx'  
df = pd.read_excel(file_path)
column_name = 'IL (dB) @1310' 
values1 = -df[column_name]

file_path2 = 'FiberConn125elite_80non_elite.xlsx'  
df = pd.read_excel(file_path2)
column_name = 'IL (dB) @1310' 
values2 = -df[column_name]

############ELITE 125-125#########
#create a histogram of 125elite-125elite using values from the pdf
file_path3 = '12SM_MT_Elite.xlsx'  
df = pd.read_excel(file_path3)
# print(df.keys())
column_name = 'x' 
bin_edges = df[column_name].to_numpy()
column_name = 'y' 
hist_values = df[column_name].to_numpy()
hist_values = hist_values[:-1]

# Normalize the histogram to create a PDF
bin_widths = np.diff(bin_edges)
pdf = hist_values / (np.sum(hist_values) * bin_widths)
# Create the cumulative distribution function (CDF)
cdf = np.cumsum(pdf * bin_widths)
# Generate random samples from the PDF (CDF method)
num_samples = 1000000  # Adjust as needed
elite125_values = np.interp(np.random.rand(num_samples), cdf, bin_edges[:-1])

########IOSMFNIT############
#create a histogram of iosmfnit
#file_path4 = 'iosmfnit.xlsx'
file_path4 = 'iosmfnit_TE.xlsx'
df = pd.read_excel(file_path4)
# print(df.keys())
column_name = 'x' 
bin_edges = df[column_name].to_numpy()
column_name = 'y' 
hist_values = df[column_name].to_numpy()
hist_values = hist_values[:-1]

# Normalize the histogram to create a PDF
bin_widths = np.diff(bin_edges)
pdf = hist_values / (np.sum(hist_values) * bin_widths)
# Create the cumulative distribution function (CDF)
cdf = np.cumsum(pdf * bin_widths)
# Generate random samples from the PDF (CDF method)
num_samples = 1000000  # Adjust as needed
iosmfnit_values = -np.interp(np.random.rand(num_samples), cdf, bin_edges[:-1])

######## INTEL CONNECTOR ############

#file_path4 = 'iosmfnit.xlsx'
file_path4 = 'Intel_Connector.xlsx'
df = pd.read_excel(file_path4)
# print(df.keys())
column_name = 'x' 
bin_edges = df[column_name].to_numpy()
column_name = 'y' 
hist_values = df[column_name].to_numpy()
hist_values = hist_values[:-1]

# Normalize the histogram to create a PDF
bin_widths = np.diff(bin_edges)
pdf = hist_values / (np.sum(hist_values) * bin_widths)
# Create the cumulative distribution function (CDF)
cdf = np.cumsum(pdf * bin_widths)
# Generate random samples from the PDF (CDF method)
num_samples = 1000000  # Adjust as needed
intel_connector_values = np.interp(np.random.rand(num_samples), cdf, bin_edges[:-1])/2


############# Directional Coupler ################
## generate normal distribution
mean = 0.50       
std_dev = .0149  
num_samples = 1000  

DC_split_ratio = np.random.normal(mean, std_dev, num_samples)

file_path5 = 'laser_mux_il.xlsx'  
df = pd.read_excel(file_path5)
column_name = 'x'
split = df[column_name]
column_name = 'y'
il = df[column_name]

laser_mux_il = -np.interp(DC_split_ratio, split, il)

############## SiN routing ######################

#create a histogram of nit_loss
file_path4 = 'nit_loss.xlsx'  
df = pd.read_excel(file_path4)
# print(df.keys())
column_name = 'x' 
bin_edges = df[column_name].to_numpy()
column_name = 'y' 
hist_values = df[column_name].to_numpy()
hist_values = hist_values[:-1]

# Normalize the histogram to create a PDF
bin_widths = np.diff(bin_edges)
pdf = hist_values / (np.sum(hist_values) * bin_widths)
# Create the cumulative distribution function (CDF)
cdf = np.cumsum(pdf * bin_widths)
# Generate random samples from the PDF (CDF method)
num_samples = 1000000  # Adjust as needed

nit_loss = np.interp(np.random.rand(num_samples), cdf, bin_edges[:-1])

#################### Tx #############################

mean = 4.7       
std_dev = .47/3  #assume 10% in doping variation/ER on rings, no NLA - carriers swept out
num_samples = 1000  

tx_tp = np.random.normal(mean, std_dev, num_samples)

################ Interleaver #################

mean = .3
std_dev = .1/3  
num_samples = 1000  

int_il = np.random.normal(mean, std_dev, num_samples)

##### pol. controller #########
mean = .4    #0.1 from controller + 0.3 from propagation
std_dev = 0.1/3   #from directional coupler variability sin^2(pi*1.06/2)
num_samples = 1000  

pol_il = np.random.normal(mean, std_dev, num_samples)

######## crr ###########
mean = .55    #0.2 + 0.15 passband ripple + 0.2 eye-degradation from propagation
std_dev = 0.1/3   #from directional coupler variability sin^2(pi*1.06/2)
num_samples = 1000  

crr_il = np.random.normal(mean, std_dev, num_samples)

######## pd ###########
mean = 0    #0.2 + 0.15 passband ripple + 0.2 eye-degradation from propagation
std_dev = 0   #from directional coupler variability sin^2(pi*1.06/2)
num_samples = 1000  

pd_il = np.random.normal(mean, std_dev, num_samples)
```
