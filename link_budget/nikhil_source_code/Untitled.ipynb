```python
#convert delta_lambda to delta_f

center_wavelength = 1310e-9 #nm
delta_lambda = 13.79e-9 #nm
c = 3e8

high_f = c/(center_wavelength-delta_lambda/2)
low_f = c/(center_wavelength+delta_lambda/2)

delta_f = high_f-low_f

print(delta_f/1e9)
```
