```python
binary_list = []
```

```python
for number in range(2**16):
    binary_value = '{0:016b}'.format(number)
    zeros = 0
    ones = 0
    for bit in binary_value:
        if bit == "0":
            zeros += 1
        if bit == "1":
            ones += 1
    if ones == zeros:
        binary_list.append(binary_value)
```

```python
switch_state_combination = []
for number in range (2**8):
    switch_state_combination.append('{0:08b}'.format(number))
```

```python
def switch_network(input_state = '1010101010101010'):
 
    output_state_1st_stage_combination = []
    for switch_state in switch_state_combination:
        output_state_1st_stage = ''
        for idd_switch in range(8):
            if switch_state[idd_switch] == '0':
                output_state_1st_stage = output_state_1st_stage + input_state[2*idd_switch:2*idd_switch+2]
            if switch_state[idd_switch] == '1':
                output_state_1st_stage = output_state_1st_stage + input_state[2*idd_switch+1] + input_state[2*idd_switch]
        output_state_1st_stage_combination.append(output_state_1st_stage)

    print(f"unique 1st stage combinations: {len(np.unique(output_state_1st_stage_combination))}")

    output_state_1st_stage_crossing_combination = []
    for input_state in np.unique(output_state_1st_stage_combination):
        even_value = ''
        odd_value = ''
        for idd_xing in range(16):
            if idd_xing%2 == 0:
                even_value += input_state[idd_xing]
            if idd_xing%2 == 1:
                odd_value += input_state[idd_xing]
        output_state_1st_stage_crossing = even_value + odd_value
        output_state_1st_stage_crossing_combination.append([output_state_1st_stage_crossing])
        
    print(f"unique 1st stage crossing combinations: {len(np.unique(output_state_1st_stage_combination))}")

    output_state_2nd_stage_combination = []
    
    for input_state in np.unique(output_state_1st_stage_crossing_combination):
        for switch_state in switch_state_combination:
            output_state_2nd_stage = ''
            for idd_switch in range(8):
                if switch_state[idd_switch] == '0':
                    output_state_2nd_stage = output_state_2nd_stage + input_state[2*idd_switch:2*idd_switch+2]
                if switch_state[idd_switch] == '1':
                    output_state_2nd_stage = output_state_2nd_stage + input_state[2*idd_switch+1] + input_state[2*idd_switch]
            output_state_2nd_stage_combination.append(output_state_2nd_stage)

    print(f"unique 2nd stage combinations: {len(np.unique(output_state_2nd_stage_combination))}")

    output_state_2nd_stage_crossing_combination = []
    for input_state in np.unique(output_state_2nd_stage_combination):
        even_value = ''
        odd_value = ''
        for idd_xing in range(8):
            if idd_xing%2 == 0:
                even_value += input_state[idd_xing]
            if idd_xing%2 == 1:
                odd_value += input_state[idd_xing]
        output_state_2nd_stage_crossing = even_value + odd_value
        even_value = ''
        odd_value = ''
        for idd_xing in range(8):
            if idd_xing%2 == 0:
                even_value += input_state[idd_xing+8]
            if idd_xing%2 == 1:
                odd_value += input_state[idd_xing+8]
        output_state_2nd_stage_crossing = output_state_2nd_stage_crossing + even_value + odd_value
        output_state_2nd_stage_crossing_combination.append([output_state_2nd_stage_crossing])

    print(f"unique 2nd stage crossing combinations: {len(np.unique(output_state_2nd_stage_crossing_combination))}")

    output_state_3rd_stage_combination = []

    for input_state in np.unique(output_state_2nd_stage_crossing_combination):
        for switch_state in switch_state_combination:
            output_state_3rd_stage = ''
            for idd_switch in range(8):
                if switch_state[idd_switch] == '0':
                    output_state_3rd_stage = output_state_3rd_stage + input_state[2*idd_switch:2*idd_switch+2]
                if switch_state[idd_switch] == '1':
                    output_state_3rd_stage = output_state_3rd_stage + input_state[2*idd_switch+1] + input_state[2*idd_switch]
            output_state_3rd_stage_combination.append(output_state_3rd_stage)
    print(f"unique 3rd stage combinations: {len(np.unique(output_state_3rd_stage_combination))}")

    output_state_3rd_stage_crossing_combination = []
    for input_state in np.unique(output_state_3rd_stage_combination):
        even_value = ''
        odd_value = ''
        for idd_xing in range(4):
            if idd_xing%2 == 0:
                even_value += input_state[idd_xing]
            if idd_xing%2 == 1:
                odd_value += input_state[idd_xing]
        output_state_3rd_stage_crossing = even_value + odd_value
        even_value = ''
        odd_value = ''
        for idd_xing in range(4):
            if idd_xing%2 == 0:
                even_value += input_state[idd_xing+4]
            if idd_xing%2 == 1:
                odd_value += input_state[idd_xing+4]
        output_state_3rd_stage_crossing = output_state_3rd_stage_crossing + even_value + odd_value
        even_value = ''
        odd_value = ''
        for idd_xing in range(4):
            if idd_xing%2 == 0:
                even_value += input_state[idd_xing+8]
            if idd_xing%2 == 1:
                odd_value += input_state[idd_xing+8]
        output_state_3rd_stage_crossing = output_state_3rd_stage_crossing + even_value + odd_value
        even_value = ''
        odd_value = ''
        for idd_xing in range(4):
            if idd_xing%2 == 0:
                even_value += input_state[idd_xing+12]
            if idd_xing%2 == 1:
                odd_value += input_state[idd_xing+12]
        output_state_3rd_stage_crossing = output_state_3rd_stage_crossing + even_value + odd_value
        output_state_3rd_stage_crossing_combination.append([output_state_3rd_stage_crossing])
    print(f"unique 3rd stage crossing combinations: {len(np.unique(output_state_3rd_stage_crossing_combination))}")

    output_state_4th_stage_combination = []

    for input_state in tqdm(np.unique(output_state_3rd_stage_crossing_combination)):
        for switch_state in switch_state_combination:
            output_state_4th_stage = ''
            for idd_switch in range(8):
                if switch_state[idd_switch] == '0':
                    output_state_4th_stage = output_state_4th_stage + input_state[2*idd_switch:2*idd_switch+2]
                if switch_state[idd_switch] == '1':
                    output_state_4th_stage = output_state_4th_stage + input_state[2*idd_switch+1] + input_state[2*idd_switch]
            output_state_4th_stage_combination.append(output_state_4th_stage)
    print(f"unique 4th stage combinations: {len(np.unique(output_state_4th_stage_combination))}")
```

```python
#Takes an input state and a switch configuration, outputs the output state
def switch_stage(input_state = '1010101010101010', switch_state='00000000'):
        output_state_1st_stage = ''
        for idd_switch in range(8):
            if switch_state[idd_switch] == '0':
                output_state_1st_stage = output_state_1st_stage + input_state[2*idd_switch:2*idd_switch+2]
            if switch_state[idd_switch] == '1':
                output_state_1st_stage = output_state_1st_stage + input_state[2*idd_switch+1] + input_state[2*idd_switch]
        return output_state_1st_stage

def crossing_stage(input_state, crossing_net):

    #crossing_net = [0,2,4,6,8,10,12,14,1,3,5,7,9,11,13,15] # the position is the output position, the value is the input position
    #crossing_net = [0,2,4,6,1,3,5,7, 8, 10,12,14, 9, 11, 13, 15]
    #crossing_net = [0,2,1,3, 4,6,5,7, 8,10,9,11, 12,14,13,15]
    reordered_state = ''.join(input_state[i] for i in crossing_net)
    return reordered_state


# input_state = '1010101010101010'
# output_state_1st_stage_combination = []
# for switch_state in switch_state_combination:
    
#     perm = switch_stage(input_state, switch_state)
#     output_state_1st_stage_combination.append(perm)
# print(f"unique 1st stage combinations: {len(np.unique(output_state_1st_stage_combination))}")
# #print(np.unique(output_state_1st_stage_combination))
```
