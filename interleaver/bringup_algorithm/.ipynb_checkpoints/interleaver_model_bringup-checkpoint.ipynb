```python
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
plt.style.use("plot_style.mplstyle")

from lmphoton.simulation import current_simulation as sim

from tqdm import tqdm

from scipy.optimize import minimize
from functools import partial
from itertools import product

from interleavers import Interleaver
```

```python
um=1e-6
wavelengths = np.linspace(1.3, 1.32, 501)*um 
no_sweep = 1001
voltages = np.linspace(0, 3.0, no_sweep)
```

# Interleaver bring up stages and steps

```python
interleaver = Interleaver()

interleaver.dc0._splitratio = 0.5
interleaver.dc1._splitratio = 0.63
interleaver.dc2._splitratio = 0.9

interleaver.XPS0.heater_voltage = 0.0
interleaver.XPS1.heater_voltage = 0.0
interleaver.XPS2.heater_voltage = 0.0
interleaver.XPS3.heater_voltage = 0.0
```

```python
def cost_calculator(wavelengths, data, channel_spacing, center_wavelength):
    
    wavelength_focus = np.linspace(center_wavelength - channel_spacing*3, center_wavelength + channel_spacing*4, 8)
    cost = 0
    for _, wavelength_focus_value in enumerate(wavelength_focus):
        if wavelength_focus_value < wavelengths[0] or wavelength_focus_value > wavelengths[-1]:
          continue
        else:    
          idd_focus = np.argmin(np.abs(wavelengths - wavelength_focus_value))
          cost += 10*np.log10(data[idd_focus])

    return cost
```

## First stage optimizer

```python
def stage_1_calculator(xps0_voltages, xps1_voltages, wavelengths, interleaver):
  bar_port_tx = []
  cross_port_tx = []

  tap_1_tx = []
  tap_2_tx = []
  idd_xps0 = 0
  idd_xps1 = 0
  tap_3_tx = []
  if xps0_voltages != []:
    idd_xps0 = 1
    for xps0_voltage in xps0_voltages:
      interleaver.XPS0.heater_voltage = xps0_voltage
      interleaver.XPS1.heater_voltage = 0.0
      interleaver.XPS2.heater_voltage = 0.0
      interleaver.XPS3.heater_voltage = 0.0
      for wavelength in wavelengths:
        sim().wavelength = wavelength
        interleaver._wavelegth = wavelength
        interleaver.recursive_update()
        s_matrix = interleaver.smatrix
        bar_port_tx.append(np.real(s_matrix * np.conj(s_matrix))[4,0])
        cross_port_tx.append(np.real(s_matrix * np.conj(s_matrix))[6,0])

        tap_1_tx.append(np.real(s_matrix * np.conj(s_matrix))[2,0])
        tap_2_tx.append(np.real(s_matrix * np.conj(s_matrix))[3,0])
        tap_3_tx.append(np.real(s_matrix * np.conj(s_matrix))[5,0])

  if xps1_voltages != []:
    idd_xps1 = 1
    for xps1_voltage in xps1_voltages:
      interleaver.XPS0.heater_voltage = 0.0
      interleaver.XPS1.heater_voltage = xps1_voltage
      interleaver.XPS2.heater_voltage = 0.0
      interleaver.XPS3.heater_voltage = 0.0
      for wavelength in wavelengths:
        sim().wavelength = wavelength
        interleaver._wavelegth = wavelength
        s_matrix = interleaver.smatrix
        bar_port_tx.append(np.real(s_matrix * np.conj(s_matrix))[4,0])
        cross_port_tx.append(np.real(s_matrix * np.conj(s_matrix))[6,0])

        tap_1_tx.append(np.real(s_matrix * np.conj(s_matrix))[2,0])
        tap_2_tx.append(np.real(s_matrix * np.conj(s_matrix))[3,0])
        tap_3_tx.append(np.real(s_matrix * np.conj(s_matrix))[5,0])

  tap_1_array = np.array(tap_1_tx).reshape(idd_xps0 * len(xps0_voltages) + idd_xps1 * len(xps1_voltages), len(wavelengths))
  tap_2_array = np.array(tap_2_tx).reshape(idd_xps0 * len(xps0_voltages) + idd_xps1 * len(xps1_voltages), len(wavelengths))
  tap_3_array = np.array(tap_3_tx).reshape(idd_xps0 * len(xps0_voltages) + idd_xps1 * len(xps1_voltages), len(wavelengths))

  bar_port_array = np.array(bar_port_tx).reshape(idd_xps0* len(xps0_voltages) + idd_xps1 * len(xps1_voltages), len(wavelengths))
  cross_port_array = np.array(cross_port_tx).reshape(idd_xps0 * len(xps0_voltages) + idd_xps1 * len(xps1_voltages), len(wavelengths))

  return tap_1_array, tap_2_array, tap_3_array, bar_port_array, cross_port_array
```

### Step 1 - Coarse sweep XPS0 and XPS1

```python
tap_1_array_step1, tap_2_array_step1, tap_3_array_step1, bar_port_array_step1, cross_port_array_step1 = stage_1_calculator(voltages, voltages, wavelengths, interleaver)

cost_array = []

for idd_xps in [0, 1]:
  for idd in tqdm(range(no_sweep)):
    cost_array.append(cost_calculator(wavelengths, tap_1_array_step1[idd + idd_xps*no_sweep,:], interleaver.channel_spacing, 1.31*um))

cost_array = np.array(cost_array)
cost_max_reached = 0
for idd_xps in [0, 1]:
  for idd in tqdm(range(no_sweep)):
    if cost_array[idd + idd_xps*no_sweep] == np.max(cost_array):
      cost_max_reached = 1
      break
  if cost_max_reached == 1:
    break
    
print(f"XPS{idd_xps} voltage: {voltages[idd]}")
```

### Step 2 - Fine sweep XPS0 and XPS1

```python
if idd_xps == 0:
  xps0_voltage_step1 = voltages[idd]
  if xps0_voltage_step1 == np.min(voltages):
    index_min = 0 
  else:
    index_min = 1
  if xps0_voltage_step1 == np.max(voltages):
    index_max = 0
  else:
    index_max = 1

  xps0_voltage_range = np.linspace(xps0_voltage_step1 - index_min*np.diff(voltages)[0], xps0_voltage_step1 + index_max*np.diff(voltages)[0], no_sweep)
  xps1_voltage_range = []
else:
  xps1_voltage_step1 = voltages[idd]
  if xps1_voltage_step1 == np.min(voltages):
    index_min = 0
  else:
    index_min = 1 

  if xps1_voltage_step1 == np.max(voltages):
    index_max = 0
  else:
    index_max = 1
  xps0_voltage_range = []
  xps1_voltage_range = np.linspace(xps1_voltage_step1 - index_min*np.diff(voltages)[0], xps1_voltage_step1 + index_max*np.diff(voltages)[0], no_sweep)

tap_1_array_step2, tap_2_array_step2, tap_3_array_step2, bar_port_array_step2, cross_port_array_step2 = stage_1_calculator(xps0_voltage_range, xps1_voltage_range, wavelengths, interleaver)

cost_array = []


for idd in tqdm(range(no_sweep)):
  cost_array.append(cost_calculator(wavelengths, tap_1_array_step1[idd,:], interleaver.channel_spacing, 1.31*um))

cost_array = np.array(cost_array)
for idd in tqdm(range(no_sweep)):
  if cost_array[idd] == np.max(cost_array):
    break

if xps0_voltage_range != []:
  xps0_voltage_opt = xps0_voltage_range[idd]
  xps1_voltage_opt = 0.0
else:
  xps0_voltage_opt = 0.0
  xps1_voltage_opt = xps1_voltage_range[idd]
    
print(f"XPS0 voltage: {xps0_voltage_opt:0.2f} V, XPS1 voltage: {xps1_voltage_opt:0.2f} V")
```

```python
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
plt.plot(wavelengths*1e9,10*np.log10(bar_port_array_step2[idd,:]), linewidth=3.0)
plt.plot(wavelengths*1e9,10*np.log10(cross_port_array_step2[idd,:]), linewidth=3.0)
plt.ylim([-20,0])
plt.xlabel("Wavelength [nm]")
plt.ylabel("Transmission [dB]")
plt.legend(["MPD","Output to TX-B"])
plt.subplot(2,1,2)
plt.plot(wavelengths*1e9,10*np.log10(tap_1_array_step2[idd,:]), linewidth=3.0)
plt.plot(wavelengths*1e9,10*np.log10(tap_3_array_step2[idd,:]), linewidth=3.0)
plt.ylim([-40,-20])
plt.xlabel("Wavelength [nm]")
plt.ylabel("Transmission [dB]")
plt.legend(["Tap_1", "Tap_2"])
plt.show()
plt.tight_layout()
```

# Second stage optimizer

```python
def stage_2_calculator(xps0_voltage_opt, xps1_voltage_opt, xps2_voltages, xps3_voltages, wavelengths, interleaver):
  bar_port_tx = []
  cross_port_tx = []

  tap_1_tx = []
  tap_2_tx = []
  idd_xps2 = 0
  idd_xps3 = 0
  tap_3_tx = []
  if xps2_voltages != []:
    idd_xps2 = 1
    for xps2_voltage in xps2_voltages:
      interleaver.XPS0.heater_voltage = xps0_voltage_opt
      interleaver.XPS1.heater_voltage = xps1_voltage_opt
      interleaver.XPS2.heater_voltage = xps2_voltage
      interleaver.XPS3.heater_voltage = 0.0
      for wavelength in wavelengths:
        sim().wavelength = wavelength
        interleaver._wavelegth = wavelength
        interleaver.recursive_update()
        s_matrix = interleaver.smatrix
        bar_port_tx.append(np.real(s_matrix * np.conj(s_matrix))[4,0])
        cross_port_tx.append(np.real(s_matrix * np.conj(s_matrix))[6,0])

        tap_1_tx.append(np.real(s_matrix * np.conj(s_matrix))[2,0])
        tap_2_tx.append(np.real(s_matrix * np.conj(s_matrix))[3,0])
        tap_3_tx.append(np.real(s_matrix * np.conj(s_matrix))[5,0])
  if xps3_voltages != []:
    idd_xps3 = 1
    for xps3_voltage in xps3_voltages:
      interleaver.XPS0.heater_voltage = xps0_voltage_opt
      interleaver.XPS1.heater_voltage = xps1_voltage_opt
      interleaver.XPS2.heater_voltage = 0.0
      interleaver.XPS3.heater_voltage = xps3_voltage
      for wavelength in wavelengths:
        sim().wavelength = wavelength
        interleaver._wavelegth = wavelength
        s_matrix = interleaver.smatrix
        bar_port_tx.append(np.real(s_matrix * np.conj(s_matrix))[4,0])
        cross_port_tx.append(np.real(s_matrix * np.conj(s_matrix))[6,0])

        tap_1_tx.append(np.real(s_matrix * np.conj(s_matrix))[2,0])
        tap_2_tx.append(np.real(s_matrix * np.conj(s_matrix))[3,0])
        tap_3_tx.append(np.real(s_matrix * np.conj(s_matrix))[5,0])

  tap_1_array = np.array(tap_1_tx).reshape(idd_xps2 * len(xps2_voltages) + idd_xps3 * len(xps3_voltages), len(wavelengths))
  tap_2_array = np.array(tap_2_tx).reshape(idd_xps2 * len(xps2_voltages) + idd_xps3 * len(xps3_voltages), len(wavelengths))
  tap_3_array = np.array(tap_3_tx).reshape(idd_xps2 * len(xps2_voltages) + idd_xps3 * len(xps3_voltages), len(wavelengths))

  bar_port_array = np.array(bar_port_tx).reshape(idd_xps2 * len(xps2_voltages) + idd_xps3 * len(xps3_voltages), len(wavelengths))
  cross_port_array = np.array(cross_port_tx).reshape(idd_xps2 * len(xps2_voltages) + idd_xps3 * len(xps3_voltages), len(wavelengths))

  return tap_1_array, tap_2_array, tap_3_array, bar_port_array, cross_port_array
```

### Step 1 - Coarse sweep XPS2 and XPS3

```python
tap_1_array_step1, tap_2_array_step1, tap_3_array_step1, bar_port_array_step1, cross_port_array_step1 = stage_2_calculator(xps0_voltage_opt, xps1_voltage_opt, voltages, voltages, wavelengths, interleaver)


cost_array = []

for idd_xps in [0, 1]:
  for idd in tqdm(range(no_sweep)):
    cost_array.append(cost_calculator(wavelengths, tap_3_array_step1[idd + idd_xps*no_sweep,:], interleaver.channel_spacing, 1.31*um))

cost_array = np.array(cost_array)
cost_max_reached = 0
for idd_xps in [0, 1]:
  for idd in tqdm(range(no_sweep)):
    if cost_array[idd + idd_xps*no_sweep] == np.max(cost_array):
      cost_max_reached = 1
      break
  if cost_max_reached == 1:
    break
    
print(f"XPS{idd_xps+2} voltage: {voltages[idd]}")
```

### Step 2 - Fine sweep XPS2 and XPS3

```python
if idd_xps == 0:
  xps2_voltage_step1 = voltages[idd]
  if xps2_voltage_step1 == np.min(voltages):
    index_min = 0 
  else:
    index_min = 1
  if xps2_voltage_step1 == np.max(voltages):
    index_max = 0
  else:
    index_max = 1

  xps2_voltage_range = np.linspace(xps2_voltage_step1 - index_min*np.diff(voltages)[0], xps2_voltage_step1 + index_max*np.diff(voltages)[0], no_sweep)
  xps3_voltage_range = []
else:
  xps3_voltage_step1 = voltages[idd]
  if xps3_voltage_step1 == np.min(voltages):
    index_min = 0
  else:
    index_min = 1 

  if xps3_voltage_step1 == np.max(voltages):
    index_max = 0
  else:
    index_max = 1
  xps2_voltage_range = []
  xps3_voltage_range = np.linspace(xps3_voltage_step1 - index_min*np.diff(voltages)[0], xps3_voltage_step1 + index_max*np.diff(voltages)[0], no_sweep)

tap_1_array_step2, tap_2_array_step2, tap_3_array_step2, bar_port_array_step2, cross_port_array_step2 = stage_2_calculator(xps0_voltage_opt, xps1_voltage_opt, xps2_voltage_range, xps3_voltage_range, wavelengths, interleaver)

cost_array = []

for idd in tqdm(range(no_sweep)):
  cost_array.append(cost_calculator(wavelengths, tap_3_array_step2[idd,:], interleaver.channel_spacing, 1.31*um))

cost_array = np.array(cost_array)
for idd in tqdm(range(no_sweep)):
  if cost_array[idd] == np.max(cost_array):
    break

if xps2_voltage_range != []:
  xps2_voltage_opt = xps2_voltage_range[idd]
  xps3_voltage_opt = 0.0
else:
  xps2_voltage_opt = 0.0
  xps3_voltage_opt = xps3_voltage_range[idd]
    
print(f"XPS2 voltage: {xps2_voltage_opt:0.2f} V, XPS3 voltage: {xps3_voltage_opt:0.2f} V")
```

```python
plt.figure(figsize=(15, 8))
plt.subplot(2,1,1)
plt.plot(wavelengths*1e9,bar_port_array_step2[idd,:], linewidth=3.0)
plt.plot(wavelengths*1e9,cross_port_array_step2[idd,:], linewidth=3.0)
plt.xlabel("Wavelength [nm]")
plt.ylabel("Transmission")
plt.legend(["MPD","Output to TX-B"])
plt.title("Grid-Search Optimized Interleaver")
plt.subplot(2,1,2)
plt.plot(wavelengths*1e9,tap_1_array_step2[idd,:], linewidth=3.0)
plt.plot(wavelengths*1e9,tap_3_array_step2[idd,:], linewidth=3.0)
plt.xlabel("Wavelength [nm]")
plt.ylabel("Transmission")
plt.legend(["Tap_1","Tap_2"])
plt.title("Grid-Search Optimized Interleaver")
plt.show()
```
