```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.linspace(0, 100, 1001)

Vturnon = 1.0

Ild = np.zeros_like(time)
Vld = np.zeros_like(time)
Vdac = np.zeros_like(time)
Virlaser = np.zeros_like(time)
Vdacoverhead = np.ones_like(time) * 0.5
Vmaxopdac = np.ones_like(time) * 1.8
Vlaser = np.ones_like(time) * 2.5

for idd_time, val_time in enumerate(time):
              if val_time <30:
                            Ild[idd_time] = 0.15 * val_time/30
              if val_time >70:
                            Ild[idd_time] = 0.15 * (100 - val_time)/30
              if val_time >=30 and val_time <=70:
                            Ild[idd_time] = 0.15
                            
Virlaser = Ild * 0.2/0.15

for idd_ild, val_ild in enumerate(Ild):
              if val_ild < 0.02:
                            Vld[idd_ild] = Vturnon * val_ild / 0.02
              if val_ild >= 0.02 :
                            Vld[idd_ild] = Vturnon + (Ild[idd_ild] - 0.02) * (1.8-Vturnon)/(0.15-0.02)
                            
Vdac = Vlaser - Vld - Virlaser #- Vdacoverhead

fig, ax1 = plt.subplots(figsize=(10, 5))
ax1.plot(time,Vlaser, linewidth = 3.0, label='Vlaser', color = 'black')
ax1.plot(time, Vld, linewidth = 2.0, label='Vld', color='red')
ax1.plot(time, Virlaser, linewidth = 1.0, label='Virlaser', linestyle = '--', color='red')
ax1.plot(time, Vdacoverhead, linewidth = 2.0, label='Vdacoverhead', color='purple')
ax1.plot(time, Vmaxopdac, linewidth = 3.0, label='Vmaxopdac', color='blue')
ax1.plot(time, Vdac, linewidth = 2.0, label='Vdac', color='orange')
ax1.set_ylim(0,3.0)
ax1.set_xlim(0, 100)
ax1.set_xlabel('Time (a.u.)')
ax1.set_ylabel('Voltage (V)')
ax1.legend(loc='upper left')
ax2=ax1.twinx()
ax2.plot(time, Ild, linewidth = 2.0, label='Ild', color='green')
ax2.legend(loc='upper right')
ax2.set_ylabel('Current (A)')
ax2.tick_params(axis='y', labelcolor='g')
ax1.grid(True)
ax2.set_ylim(0, 0.3)
fig.tight_layout()
```
