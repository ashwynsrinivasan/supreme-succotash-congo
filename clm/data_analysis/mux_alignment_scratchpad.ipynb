```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sci
```

```python
background = 175 # uA
```

```python
current_0 = 25 

current_sweep = np.linspace(10, 40, num=51)

mux_tf =  np.exp(-(current_sweep - current_0)**2 / 20)
plt.plot(current_sweep,mux_tf)
plt.grid(True)
```

```python
plt.plot(current_sweep, mux_tf)
```

```python
plt.plot(current_sweep, current_sweep)
```

```python
plt.plot(current_sweep, mux_tf * current_sweep / np.max(current_sweep * mux_tf), label='Signal multiplied ')
plt.plot(current_sweep, mux_tf, label='mux_tf')
plt.legend()
```

```python
mux_signal = mux_tf * background *current_sweep/15 + background
plt.plot(current_sweep, mux_tf, label='mux_tf')
plt.plot(current_sweep, mux_signal/np.max(mux_signal), label='mpd_raw_signal')
normalization = mux_signal/current_sweep

# dither_normalization = np.diff(mux_signal)/np.diff(current_sweep)
# dither_normalization_2 = np.diff(mux_signal)/ current_sweep[1:]

plt.plot(current_sweep, normalization / np.max(normalization), label='only current ratios')
# plt.plot(current_sweep[1:], dither_normalization / np.max(dither_normalization), label='dither current ratios')
# plt.plot(current_sweep[1:], dither_normalization_2 /np.max(dither_normalization_2), label='dither current ratios 2')
# plt.xlim(20, 30)
plt.legend()
plt.grid(True)
```

```python
plt.plot(current_sweep, background * np.ones_like(current_sweep), label='background')
plt.plot(current_sweep, mux_tf * background *current_sweep/15 + background, label='mux_signal')
plt.plot(current_sweep, mux_tf * background *current_sweep/30 + background, label='mux_signal_2')
plt.legend()
```
