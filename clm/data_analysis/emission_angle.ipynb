```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n1_wg = 1.6
n2_air = 1

emission_angle = 2.5
```

```python
theta_deg = np.linspace(0,10, num=101)
theta_rad = theta_deg/180 * np.pi
```

```python
phi_rad = np.arcsin(n1_wg * np.sin(theta_rad) / n2_air)
phi_deg = phi_rad / np.pi * 180
```

```python
plt.figure(figsize=(12,5))
plt.subplot(121)
plt.plot(theta_deg, phi_deg, linewidth = 2.0)
plt.xlabel(r"$\Theta$ [deg]")
plt.ylabel(r"$\Phi$ [deg]")
plt.grid(True)
plt.subplot(122)
plt.plot(theta_deg, - phi_deg + theta_deg, linewidth = 2.0)
plt.plot(theta_deg, - 2.5*np.ones_like(theta_deg), linewidth = 2.0, linestyle = '--', label="PRD spec rev0.5")
plt.legend()
plt.grid(True)
plt.xlabel(r"$\Theta$ [deg]")
plt.ylabel(r"Emission angle [deg]")
```
