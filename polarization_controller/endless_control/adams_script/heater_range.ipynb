```python
import numpy as np
import matplotlib.pyplot as plt

N = 13
vref = 1.65
offset = 0.15

code_p = np.arange(2**N)
code_p_min = int(offset**2 / vref**2 * 2**(N))
code_v = np.sqrt(code_p) * 2**(N/2)
voltage = code_v / 2**N * (vref)


code_v_min = np.sqrt(code_p_min) * 2**(N/2)
voltage_min = code_v_min / 2**N * vref

print(f'calculated min_code = {code_p_min} for voltage {voltage_min} V')

fig, ax = plt.subplots(2, 1)
ax[0].plot(code_p, code_v)
ax[0].axvline(x=code_p_min, color='red', linestyle='--')
ax[0].axhline(y=code_v_min, color='red', linestyle='--')
ax[0].set_title('HDAC code vs. Algorithm code')
ax[0].set_ylabel('HDAC code')
ax[1].plot(code_p, voltage)
ax[1].axvline(x=code_p_min, color='red', linestyle='--')
ax[1].axhline(y=voltage_min, color='red', linestyle='--')
ax[1].set_title('Output voltage vs. Algorithm code')
ax[1].set_ylabel('Voltage (V)')
ax[1].set_xlabel('Alg. Code')

plt.subplots_adjust(hspace=0.5)

```
